```python
import os, codecs, spacy
sp = spacy.load('en_core_web_sm')
```

```python
def cleanword(w):
    w = w.lower()

    if not w.isalnum() and len(w) == 1:
        return w

    while w and not w[0].isalnum():
        w = w[1:]
    while w and not w[-1].isalnum():
        w = w[:-1]
    return w
```

```python
def dir2files(somedirectory, path = True):
    if path == False:
        files = os.listdir(somedirectory)
    else:
        files = os.listdir(somedirectory)
        files = [os.path.join(somedirectory, f) for f in files]

    for f in files[:]:  # go to next object
        if f[-4:] != ".txt":
            files.remove(f)
    return files
```

```python
def file2token(file_name):
    file = codecs.open(file_name, encoding = "utf8")
    text = file.read()

    short_story_text = sp(text)

    return short_story_text
```

```python
def write2output(text, ofn, extra_col = "", sep = "\t"):
    with open(ofn, 'w') as output_file:
        for word in text:
            write_str = (cleanword(word.text), word.pos_, word.tag_, str(spacy.explain(word.tag_)), 
                         str(extra_col), '\n')


            write_str = sep.join(write_str)
            
            if write_str[0] == "\"":
                write_str = write_str.replace("\"", "||")    # quotation marks appear as ||
            if word.pos_ == 'SPACE':
                continue

            output_file.write(write_str)

        output_file.close()
        print("Wrote " + extra_col + " to the file " + ofn)
```

```python
# LIKE MAIN
def create_output_file():
    file_list = dir2files(sdir)

    for file in file_list:
        short_story = file2token(file)

        ofn = "/Users/eunjilee/Desktop/cesta/short_stories_project/pos_tagged/" + file[-8:-4] + "_postagged" + ".tsv"
        print(ofn)

        write2output(short_story, ofn, extra_col = file[-8:])
```
